# Introducción a la Programación Orientada a Objetos (POO)

¿Qué es la Programación Orientada a Objetos?
La Programación Orientada a Objetos (POO) es un paradigma de programación que utiliza "objetos" y "clases". Es una forma de organizar y estructurar el código de manera que sea más fácil de entender, mantener y reutilizar.



## Conceptos básicos
1. Clase
  * Una clase es como un plano o molde que define las características y comportamientos de un objeto.
  * Las clases contienen atributos (datos) y métodos (funciones).

2. Objetos
  * Un objeto es una instancia de una clase.
  * Cada objeto puede tener diferentes valores para sus atributos, pero comparte los mismos métodos definidos en la clase.

3. Atributos
  * Los atributos son variables que pertencen a una clase u objeto.
  * Definen las propiedades o caracaterísticas del objeto.

4. Métodos
  * Los métodos son funciones que pertenencen a una clase y pueden ser llamados por sus objetos.
  * Definen los comportamientos del objeto.
